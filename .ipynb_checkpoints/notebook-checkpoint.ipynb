```python
# Cell 1: Imports + load dataset
import numpy as np
import pandas as pd
import os

# sklearn imports we'll need later (import now so environment errors show early)
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report

# Print working directory and files (helpful to ensure your CSV is in the right place)
print("Working dir:", os.getcwd())
print("Files in working dir (first 20):")
print("\n".join(os.listdir('.')[:20]))

# Load CSV - replace path if file is in a different folder
csv_path = "diabetes_binary_5050split_health_indicators_BRFSS2015.csv"

df = pd.read_csv(csv_path)
print("\nDataset loaded successfully.")

# Quick inspections
print("\nShape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nTarget value counts (Diabetes_binary):")
print(df['Diabetes_binary'].value_counts(dropna=False))

print("\nFirst 5 rows:")
display(df.head())

print("\nData types and missing values per column:")
missing = df.isnull().sum()
dtypes = df.dtypes
info_df = pd.DataFrame({'dtype': dtypes, 'missing_count': missing, 'missing_pct': missing / len(df)})
display(info_df.sort_values('missing_count', ascending=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the figure
plt.figure(figsize=(12, 8))
corr = df.corr()

# Plot heatmap
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap of Features", fontsize=14)
plt.show()

# Let's also check the correlation of each feature with the target variable
target_corr = corr['Diabetes_binary'].sort_values(ascending=False)
print("\nCorrelation of each feature with Diabetes_binary:\n")
print(target_corr)
```

```python
# Cell 2: Separate features and target, then train-test split

# Separate features and label
X = df.drop(columns=['Diabetes_binary'])
y = df['Diabetes_binary']

# Train-test split (stratify keeps same class balance in both sets)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("âœ… Data split done successfully")
print("Training set:", X_train.shape, "Testing set:", X_test.shape)
print("Class distribution in training set:\n", y_train.value_counts(normalize=True))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# 1ï¸âƒ£ Class balance visualization
plt.figure(figsize=(6,4))
sns.countplot(x='Diabetes_binary', data=df, palette='coolwarm')
plt.title('Target Variable Distribution (Diabetes_binary)')
plt.xlabel('Diabetes_binary (0 = No Diabetes, 1 = Diabetes)')
plt.ylabel('Count')
plt.show()

# 2ï¸âƒ£ Correlation heatmap (top features only for readability)
plt.figure(figsize=(10,8))
corr = df.corr()
sns.heatmap(corr, annot=False, cmap='coolwarm', center=0)
plt.title('Feature Correlation Heatmap')
plt.show()

# 3ï¸âƒ£ Example distribution of BMI
plt.figure(figsize=(6,4))
sns.histplot(df['BMI'], bins=30, kde=True, color='steelblue')
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

```python
# Cell 3: Baseline model (no PCA)

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix, classification_report

# Create baseline pipeline
pipe_baseline = Pipeline([
    ('scaler', StandardScaler()),
    ('clf', LogisticRegression(max_iter=1000, random_state=42))
])

# Fit and evaluate
pipe_baseline.fit(X_train, y_train)
y_pred = pipe_baseline.predict(X_test)

# Metrics
acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

# Print results
print("âœ… Baseline Model (No PCA)")
print(f"Accuracy: {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall: {rec:.4f}")
print(f"F1 Score: {f1:.4f}")
print("\nConfusion Matrix:\n", cm)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit on training data and transform both train and test
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("âœ… Data scaling done successfully.")
print("Scaled training shape:", X_train_scaled.shape)
print("Scaled testing shape:", X_test_scaled.shape)
```

```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

# Apply PCA without reducing dimensionality (just to analyze variance)
pca = PCA()
pca.fit(X_train_scaled)

# Calculate cumulative explained variance
cum_var = np.cumsum(pca.explained_variance_ratio_)

# Plot cumulative variance
plt.figure(figsize=(8,5))
plt.plot(range(1, len(cum_var)+1), cum_var, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA Explained Variance Analysis')
plt.grid(True)
plt.show()

# Print variance percentages for first few components
for i, var in enumerate(cum_var[:10], 1):
    print(f"PC{i}: {var:.4f} cumulative variance")

import matplotlib.pyplot as plt
import numpy as np

# Fit PCA on full scaled training data
pca_full = PCA().fit(X_train_scaled)
cum_var = np.cumsum(pca_full.explained_variance_ratio_)

plt.figure(figsize=(8,5))
plt.plot(range(1, len(cum_var)+1), cum_var, marker='o', color='royalblue')
plt.axhline(y=0.85, color='green', linestyle='--', label='85% variance')
plt.axhline(y=0.90, color='red', linestyle='--', label='90% variance')
plt.title('PCA Cumulative Explained Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Variance Ratio')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Check how many components are needed to explain ~85â€“90% variance
import numpy as np

pca = PCA()
pca.fit(X_train_scaled)

cum_var = np.cumsum(pca.explained_variance_ratio_)

# Find the smallest number of components where cumulative variance >= 0.90
n_components_85 = np.argmax(cum_var >= 0.85) + 1
n_components_90 = np.argmax(cum_var >= 0.90) + 1

print(f"Components needed for 85% variance: {n_components_85}")
print(f"Components needed for 90% variance: {n_components_90}")

# Optional: show a few nearby values
for i, val in enumerate(cum_var[:20], start=1):
    print(f"PC{i}: {val:.4f}")
```

```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix

def train_pca_model(n_components):
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train_scaled)
    X_test_pca = pca.transform(X_test_scaled)

    model = LogisticRegression(max_iter=1000, random_state=42)
    model.fit(X_train_pca, y_train)
    y_pred = model.predict(X_test_pca)

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    print(f"\nðŸ”¹ PCA with {n_components} components")
    print(f"Accuracy: {acc:.4f}, F1 Score: {f1:.4f}")
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("\nClassification Report:\n", classification_report(y_test, y_pred))
    return acc, f1

# Train for PCA(15), PCA(16), PCA(17)
results = {}
for n in [15, 16, 17]:
    results[n] = train_pca_model(n)
```

```python
from sklearn.model_selection import GridSearchCV

# Use PCA configuration (16 components)
pca_best = PCA(n_components=16)
X_train_pca = pca_best.fit_transform(X_train_scaled)
X_test_pca = pca_best.transform(X_test_scaled)

# Logistic Regression hyperparameter grid
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['lbfgs', 'liblinear'],
    'penalty': ['l2']
}

grid = GridSearchCV(LogisticRegression(max_iter=1000, random_state=42),
                    param_grid,
                    cv=5,
                    scoring='f1',
                    n_jobs=-1)

grid.fit(X_train_pca, y_train)

print("âœ… Grid Search Completed")
print("Best Parameters:", grid.best_params_)
print("Best Cross-Validation F1 Score:", grid.best_score_)

# Evaluate the best model on test data
best_model = grid.best_estimator_
y_pred_best = best_model.predict(X_test_pca)

print("\nTest Accuracy:", accuracy_score(y_test, y_pred_best))
print("Test F1 Score:", f1_score(y_test, y_pred_best))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_best))
```

```python
### ðŸ”¹ Manual Hyperparameter Tuning for Logistic Regression

from sklearn.metrics import f1_score, accuracy_score
import numpy as np

C_values = [0.01, 0.1, 1, 10]
manual_results = []

for c in C_values:
    model = LogisticRegression(C=c, penalty='l2', solver='lbfgs', max_iter=1000, random_state=42)
    model.fit(X_train_pca, y_train)
    y_pred = model.predict(X_test_pca)
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    manual_results.append((c, acc, f1))

manual_df = pd.DataFrame(manual_results, columns=['C Value', 'Accuracy', 'F1 Score'])
print("âœ… Manual Hyperparameter Tuning Results:")
display(manual_df)
```

```python
#Cell â€” MinMaxScaler + PCA(16)

from sklearn.preprocessing import MinMaxScaler

# 1ï¸âƒ£ Apply MinMax Scaling instead of StandardScaler
scaler_mm = MinMaxScaler()
X_train_mm = scaler_mm.fit_transform(X_train)
X_test_mm = scaler_mm.transform(X_test)

# 2ï¸âƒ£ Apply PCA (16 components) after MinMax scaling
pca_mm = PCA(n_components=16, random_state=42)
X_train_pca_mm = pca_mm.fit_transform(X_train_mm)
X_test_pca_mm = pca_mm.transform(X_test_mm)

# 3ï¸âƒ£ Train Logistic Regression model
logreg_mm = LogisticRegression(max_iter=1000, solver='lbfgs')
logreg_mm.fit(X_train_pca_mm, y_train)

# 4ï¸âƒ£ Evaluate
y_pred_mm = logreg_mm.predict(X_test_pca_mm)

acc_mm = accuracy_score(y_test, y_pred_mm)
f1_mm = f1_score(y_test, y_pred_mm)
print(f"ðŸ”¹ Logistic Regression (MinMaxScaler + PCA 16)\nAccuracy: {acc_mm:.4f}, F1 Score: {f1_mm:.4f}\n")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_mm))
print("\nClassification Report:\n", classification_report(y_test, y_pred_mm))
```

```python
# Calculate correlation of each feature with the target
corr_with_target = df.corr()['Diabetes_binary'].abs().sort_values(ascending=False)

# Select features with correlation > 0.05 (tunable threshold)
selected_features = corr_with_target[corr_with_target > 0.05].index.tolist()
selected_features.remove('Diabetes_binary')  # remove target column

print("Selected features based on correlation > 0.05:\n", selected_features)

# Prepare data
X_train_fs = X_train[selected_features]
X_test_fs = X_test[selected_features]

# Scale selected features
scaler_fs = StandardScaler()
X_train_fs_scaled = scaler_fs.fit_transform(X_train_fs)
X_test_fs_scaled = scaler_fs.transform(X_test_fs)

# Train Logistic Regression on selected features
logreg_fs = LogisticRegression(max_iter=1000, random_state=42)
logreg_fs.fit(X_train_fs_scaled, y_train)

# Evaluate
y_pred_fs = logreg_fs.predict(X_test_fs_scaled)
acc_fs = accuracy_score(y_test, y_pred_fs)
f1_fs = f1_score(y_test, y_pred_fs)

print(f"\nðŸ”¹ Logistic Regression (Feature Selection >0.05)")
print(f"Accuracy: {acc_fs:.4f}, F1 Score: {f1_fs:.4f}")
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_fs))
```
